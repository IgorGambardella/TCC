```python
from CoolProp.CoolProp import PropsSI as COOLPROP
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import get_test_data

#Listas
registro_fluidos = []
COP_lista = []
lista_T4 = []
Wcomp_total_lista = []
cont = []

#Lista de Temperaturas
for T4 in range (-3000 + 27315, 400 + 27315, 100):
    T4 = T4 / 100
    cont.append(T4)
    
#Lista de caractersticas para as curvas do grfico
lista_cor = ['r','g','b', 'c', 'm']
lista_ls = ['-', '-.', ':']
caracteristicas = []
for ls in lista_ls:
    for cor in lista_cor:
        caracteristicas.append((cor, ls))

#Dados do problema
Pfrig = 120 * 1000 #[W]
T1 = -80 + 273.15 #[K]
T10 = 28 + 273.15 #[K]
N = 0.7 #Eficincia isentrpica
fluid_high_lista = ['R134a', 'R717', 'R290', 'R600', 'R404a']
fluid_low_lista = ['R23', 'R170', 'NitrousOxide']
drop_suc = 12000 #[Pa]
drop_desc = 24000 #[Pa]
efet = 0.7 #efetividade do trocador intermedirio
DeltaT = 4 #[C]

#variao dos refrigerantes e do coeficiente politrpico
for fluid_low in fluid_low_lista:
    for fluid_high in fluid_high_lista:
        if fluid_low == 'NitrousOxide':
            par_fluido = ('xido nitroso', fluid_high)    
        else:
            par_fluido = (fluid_low, fluid_high)
        registro_fluidos.append(par_fluido)
        #Presso crtica do fluido ciclo de alta presso
        P_crit_high = COOLPROP('PCRIT', fluid_high)
        for T4 in cont:
            #Ciclo de baixa presso
            cp1 = COOLPROP('CP0MASS', 'T', T1, 'Q', 1, fluid_low)
            cp4 = COOLPROP('CP0MASS', 'T', T4, 'Q', 0, fluid_low)
            P1 = COOLPROP('P', 'T', T1, 'Q', 1, fluid_low)
            P4 = COOLPROP('P', 'T', T4, 'Q', 0, fluid_low)
            h1 = COOLPROP('H', 'T', T1, 'Q', 1, fluid_low)
            h4 = COOLPROP('H', 'T', T4, 'Q', 0, fluid_low)
            P5 = P4
            P2 = P1
            
            #Comparao entre os calores especficos nos pontos 4 e 1
            if cp1 < cp4:
                
                T2 = efet * (T4 - T1) + T1
                h2 = COOLPROP('H', 'T', T2, 'P', P2, fluid_low)
                hw = h2
                Pw = P2 - drop_suc
                Tw = COOLPROP('T', 'H', hw, 'P', Pw, fluid_low)
                Px = Pw
                Tx = Tw + DeltaT
                hx = COOLPROP('H', 'P', Px, 'T', Tx, fluid_low)
                sx = COOLPROP('S', 'P', Px, 'T', Tx, fluid_low)
            
                P3 = P4
                Py = P3 + drop_desc
                sy_ideal = sx
                hy_ideal = COOLPROP('H', 'P', Py, 'S', sy_ideal, fluid_low)
                hy_real = (hy_ideal - hx)/N + hx
                h3 = hy_real
            
                h5 = h4 + h1 - h2
                T5 = COOLPROP('T', 'H', h5, 'P', P5, fluid_low)
                h6 = h5
                
            else:
                T5 = T4 - efet * (T4 - T1)
                h5 = COOLPROP('H', 'T', T5, 'P', P5, fluid_low)
                h2 = h4 + h1 - h5
                T2 = COOLPROP('T', 'H', h2, 'P', P2, fluid_low)
                
                hw = h2
                Pw = P2 - drop_suc
                Tw = COOLPROP('T', 'H', hw, 'P', Pw, fluid_low)
                Px = Pw
                Tx = Tw + DeltaT
                hx = COOLPROP('H', 'P', Px, 'T', Tx, fluid_low)
                sx = COOLPROP('S', 'P', Px, 'T', Tx, fluid_low)
                P3 = P4
                Py = P3 + drop_desc
                sy_ideal = sx
                hy_ideal = COOLPROP('H', 'P', Py, 'S', sy_ideal, fluid_low)
                hy_real = (hy_ideal - hx)/N + hx
                h3 = hy_real
            
                h6 = h5
                
            m_flow_low = Pfrig / (h1 - h6)
            w_low = hy_real - hx
            Wcomp_low = m_flow_low * w_low
            
            
            #Presso crtica do fluido do ciclo de alta presso
            P_crit_high = COOLPROP('PCRIT', fluid_high)
            h10 = COOLPROP('H', 'T', T10, 'Q', 0, fluid_high)
            P10 = COOLPROP('P', 'T', T10, 'Q', 0, fluid_high)

            T12 = T4 - 5
            #Q<>0, mas como o refrigerante se encontra em uma fase saturada a presso  a mesma para qualquer Q com T=T10
            P12 = COOLPROP('P', 'Q', 0, 'T', T12, fluid_high) 
            P7 = P12
            T7 = COOLPROP('T', 'P', P7, 'Q', 1, fluid_high)
            P10 = COOLPROP('P', 'T', T10, 'Q', 0, fluid_high)
            cp7 = COOLPROP('CP0MASS', 'T', T7, 'Q', 1, fluid_high)
            cp10 = COOLPROP('CP0MASS', 'T', T10, 'Q', 0, fluid_high)
            h7 = COOLPROP('H', 'T', T7, 'Q', 1, fluid_high)
            h10 = COOLPROP('H', 'T', T10, 'Q', 0, fluid_high)
            P8 = P7
            P11 = P10

        #Comparao entre os calores especficos nos pontos 10 e 7
            if cp7 < cp10:
                T8 = efet * (T10 - T7) + T7
                h8 = COOLPROP('H', 'T', T8, 'P', P8, fluid_high)
                ha = h8
                Pa = P8 - drop_suc
                Ta = COOLPROP('T', 'H', ha, 'P', Pa, fluid_high)
                Pb = Pa
                Tb = Ta + DeltaT
                hb = COOLPROP('H', 'P', Pb, 'T', Tb, fluid_high)
                sb = COOLPROP('S', 'P', Pb, 'T', Tb, fluid_high)
            
                P9 = P10
                Pc = P9 + drop_desc
                sc_ideal = sb
                hc_ideal = COOLPROP('H', 'P', Pc, 'S', sc_ideal, fluid_high)
                hc_real = (hc_ideal - hb)/N + hb
                h9 = hc_real
            
                h11 = h10 + h7 - h8
                T11 = COOLPROP('T', 'H', h11, 'P', P11, fluid_high)
                h12 = h11
            else:
                T11 = T10 - efet * (T10 - T7)
                h11 = COOLPROP('H', 'T', T11, 'P', P11, fluid_high)
                h8 = h10 + h7 - h11
                T8 = COOLPROP('T', 'H', h8, 'P', P8, fluid_high)
                ha = h8
                Pa = P8 - drop_suc
                Ta = COOLPROP('T', 'H', ha, 'P', Pa, fluid_high)
                Pb = Pa
                Tb = Ta + DeltaT
                hb = COOLPROP('H', 'P', Pb, 'T', Tb, fluid_high)
                sb = COOLPROP('S', 'P', Pb, 'T', Tb, fluid_high)
            
                P9 = P10
                Py = P9 + drop_desc
                sc_ideal = sb
                hc_ideal = COOLPROP('H', 'P', Pc, 'S', sc_ideal, fluid_high)
                hc_real = (hc_ideal - hb)/N + hb
                h7 = hc_real
            
                h12 = h11
            
            #Registro dos graus de subresfriamento e superaquecimento
            grau_subresf = T10 - T11
            grau_superaq = T8 - T7
            #print('O grau de subresfriamento ({}, {})  igual a {:0.2f}C.'.format(fluid_low,fluid_high,grau_subresf))
            #print('O grau de superaquecimento ({}, {})  igual a {:0.2f}C.'.format(fluid_low,fluid_high,grau_superaq))

            #Calor trocado no trocador central
            Qcond_low =  m_flow_low * (h3 - h4)
            #Ciclo de alta presso
            Qevap_high = Qcond_low
            m_flow_high = Qevap_high / (h7 - h12)
            #Verificao da presso crtica
            if P_crit_high > P9:
                lista_T4.append(T4)
                #Trabalho especfico do compressor do ciclo de alta presso
                w_high = hc_real - hb
                #Trabalho no compressor do ciclo de alta presso
                Wcomp_high = m_flow_high * w_high
                #Trabalho total do ciclo
                Wcomp_total = Wcomp_low + Wcomp_high
                Wcomp_total_lista.append(Wcomp_total / 1000)
                #COP
                COP = Pfrig / Wcomp_total
                COP_lista.append(COP)
                print("A", fluid_high)
                print("B", fluid_low)
                print("T4", T4)
                print("C", COP)
            else:
                print ('A presso P9 = {:0.2f} kPa (T4 = {}, fluido inf = {}, fluido sup = {})  maior do que a ' 
               'presso crtica do {} logo o ponto no foi includo no grfico'.format(P9 / 1000,T4,fluid_low,
                                                                                               fluid_high, fluid_high))

if len(lista_T4) != 0:
    plt.rc('xtick', labelsize = 13)
    plt.rc('ytick', labelsize = 13)
    plt.figure(figsize = [13, 13], dpi = 300)
    for i in range (0, len(COP_lista), len(cont)):
        eixo_x = []
        eixo_y = []
        for j in range (i, i + len(cont), 1):
            eixo_x.append(lista_T4[j])
            eixo_y.append(COP_lista[j])
        plt.plot(eixo_x, eixo_y, color = caracteristicas[int(i/len(cont))][0], 
                 linestyle = caracteristicas [int(i/len(cont))][1], 
                 label=registro_fluidos[int(i/len(cont))])
        plt.xlabel('T4', size = 13)
        plt.ylabel('COP', size = 13)
    plt.legend(prop={'size' : 13})
    plt.show()
else:
    print('Todos os pontos so invlidos')     
        
```
